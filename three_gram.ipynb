```
from numpy import array
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.callbacks import CSVLogger
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GRU,Embedding
```

```
import json

with open('/content/three_gram_dataset.json','r',encoding='utf-8') as file:
    lines = json.load(file)

print(type(lines))
print(lines[1000])
```

```
len(lines)
```

```
tokenizer = Tokenizer()
tokenizer.fit_on_texts(lines)
sequences = tokenizer.texts_to_sequences(lines)
vocab_size = len(tokenizer.word_index) + 1
```

```
sequences = array(sequences)
sequences[771]
```

```
c=1
h=0
while(c!=0):
    limit = sequences.size
    d= []
    c=0
    for i in range(0, limit):
        l=len(sequences[i])
        if(l!=4):
            print(sequences[i])
            print(lines[i])

            sequences = np.delete(sequences, i, 0)
            c=1
            print("error at:",i)
            print("Deleting at: _______________", i)
            h=h+1
            break

print("total deletion: ", h)
```

```
print(sequences.size)
```

```
sequences = np.array(sequences.tolist())
sequences
```

```
sequences.shape
```

```
sequences = array(sequences)
x,y = sequences[:,:-1], sequences[:,-1]
y = to_categorical(y, num_classes=vocab_size)
seq_length = x.shape[1]
```

```
seq_length
```

```
model = Sequential()
model.add(Embedding(vocab_size, 50, input_length=seq_length))
model.add(GRU(100, return_sequences=True))
model.add(GRU(100))
model.add(Dense(vocab_size, activation='softmax'))
print(model.summary())

model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
csv_logger = CSVLogger('training_3_1.log')
model.fit(x, y, batch_size = 500, epochs =1000, callbacks = [csv_logger])
```

```
from pickle import dump

model.save('three_gram_model.h5')
dump(tokenizer, open(r'three_gram_tokenizer.pkl','wb'))
```
